```python
import os
import sys
import pandas as pd
sys.path.insert(0, "../../")
from src.text.epu import EPU
```

```python
target_dir = sys.path[0] + "data/text/tonga/"
```

```python
mtg = EPU(target_dir+"matangi_news.csv")
mtg.get_epu_category()
mtg_epu = mtg.get_epu_stats(cutoff="2020-01-01").fillna(0)
```

```python
scaling_factor = 100/mtg_epu[mtg_epu.date>="2020-01-01"]["z_score"].mean()
mtg_epu["epu"] = mtg_epu["z_score"] * scaling_factor
mtg_epu.plot(x="date", y="epu")
```
